```python
def GINI_split_attribute(D,A):
    
    if A=='Sex':
        left=[]
        right=[]
        for i in range(len(D)):
            if int(D[i][0])==0:
                left.append(D[i])
            else :
                right.append(D[i])
    elif A=='Pclass 1 2':
        left=[]
        right=[]
        for i in range(len(D)):
            if int(D[i][1])<3:
                left.append(D[i])
            else :
                right.append(D[i])
    elif A=='Pclass 2 3':
        left=[]
        right=[]
        for i in range(len(D)):
            if int(D[i][1])>1:
                left.append(D[i])
            else :
                right.append(D[i])
    
    elif A=='Embarked 0 1':
        left=[]
        right=[]
        for i in range(len(D)):
            if int(D[i][2])<2:
                left.append(D[i])
            else :
                right.append(D[i])
    
    elif A=='Embarked 1 2':
        left=[]
        right=[]
        for i in range(len(D)):
            if int(D[i][2])>0:
                left.append(D[i])
            else :
                right.append(D[i])
    
    if len(left)==0 or len(right)==0:
        return(0)
    
    #we first take care of left side
    L,R=[],[] #we calculate the GINI split
        
    for i in range(len(left)):
        if int(left[i][3])==0:
            L.append(left[i])
        else :
            R.append(left[i])
    left=L+R
    gini_left=1-(len(L)/len(left))**2-(len(R)/len(left))**2
    
        
        
        #now we take care of right side
        
    L,R=[],[] 
        
    for i in range(len(right)):
        if int(right[i][3])==0:
            L.append(right[i])
        else :
            R.append(right[i])
    right=L+R
    gini_right=1-(len(L)/len(right))**2-(len(R)/len(right))**2
    
        
    GINI=(len(left)/len(D))*gini_left + (len(right)/len(D))*gini_right
    return([GINI,left,right,A])
    
```

```python
def GINI_split(D):

    pclass1=GINI_split_attribute(D,'Pclass 1 2')
    pclass2=GINI_split_attribute(D,'Pclass 2 3')
    em0=GINI_split_attribute(D,'Embarked 0 1')
    em1=GINI_split_attribute(D,'Embarked 1 2')
    L=[pclass1,pclass2,em0,em1]
    M=[]
    for i in range(len(L)):  #here we keep only the attributes for which we have been able to split
        if L[i]!=0:
            M.append(L[i])
    ginis=[x[0] for x in M]
    m=min(ginis)
    return(M[ginis.index(m)])
```

```python
def GINI_node(Node):
    L=[]
    R=[]
    for i in range(len(Node)): #Survived
        if int(Node[i][3])==0:
            L.append(Node[i])
        else :
            R.append(Node[i])
    GINI=1-(len(L)/len(Node))**2-(len(R)/len(Node))**2
    return(GINI)
```

```python
def is_a_leaf(D,minNum): #this function calculates if, yes or no, D is a leaf according to the GINI criterion, of minNum, and the fact that if all the records are equal except the last one, then we have to deal with a leaf
    
    if GINI_node(D)==0:
        return([True,0])
    if len(D)<minNum :
        return([True,1])
    sex=D[0][0]
    pclass=D[0][1]
    em=D[0][2]
    c=0
    for i in range(len(D)): 
        if D[i][0]!=sex or D[0][1]!=pclass or D[i][2]!=em:
            c+=1
    if c==0 and len(D)>=minNum and GINI_node(D)>0:
        return([True,2])
    else:
        return([False,0])
```

```python
def majority_class(D):
    L=[]
    R=[]
    for i in range(len(D)):
        if int(D[i][3])==0 :
            L.append(D[i][3])
        else:
            R.append(D[i][3])
    if len(L)>len(R):
        return(0)
    else:
        return(1)
```

```python
def root(D):
    L=[]
    R=[]
    for i in range(len(D)): #we split according to Feature Sex<1
        if int(D[i][0])==0:
            L.append(D[i])
        else:
            R.append(D[i])
    return(['root',D,GINI_node(D),'Sex<1',1,0],L,R)
```

```python
def BuildDecisionTree_from_root(D,minNum,n,level,Tree,d):
    
    if is_a_leaf(D,minNum)[0] is True:
        if is_a_leaf(D,minNum)[1]==2:                        #if we differentiate expresses the class differently according to the condition of being a leaf
            level+=1
            L=['leaf',D, GINI_node(D),majority_class(D),n,level]
            Tree.append(L)
        elif is_a_leaf(D,minNum)[1]==1:
            level+=1
            L=['leaf',D, GINI_node(D),d,n,level]
            Tree.append(L)
        elif is_a_leaf(D,minNum)[1]==0:
            level+=1
            L=['leaf',D, GINI_node(D),majority_class(D),n,level]
            Tree.append(L)
    
    else :
        level+=1
        v1,v2=GINI_split(D)[1],GINI_split(D)[2]
        L=['node',D,GINI_node(D),GINI_split(D)[3],n,level]
        Tree.append(L)
        BuildDecisionTree_from_root(v1,minNum,2*n,level,Tree,d)
        BuildDecisionTree_from_root(v2,minNum,2*n+1,level,Tree,d)
    return(Tree)
```

```python
def max_level(Tree):
    L=[]
    for i in range(len(Tree)):
        L.append(Tree[i][5])
    return(max(L))
```

```python
def get_level(i,Tree):
    L=[]
    for j in range(len(Tree)):
        if Tree[j][5]==i:
            L.append(Tree[j])
    return(L)
```
